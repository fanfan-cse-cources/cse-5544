```python
import pandas as pd


"""
1. Read the energy data 
2. Find total generation (TYPE OF PRODUCER is 'Total Electric Power Industry' \
and ENERGY SOURCE is 'Total') data records from 2021/01 to 2021/12 (US-TOTAL for STATE is excluded)
"""

energy = pd.read_csv("./dataset/electricity_generation.csv", index_col = 0)
energy = energy.rename(columns={ 'GENERATION (Megawatthours)': 'GENERATION(Mwh)' })
energy_2021_original = energy.loc[(energy['YEAR'] == 2021)].copy()
energy_2021 = energy_2021_original.loc[(energy_2021_original['TYPE OF PRODUCER'] == 'Total Electric Power Industry') &
                                (energy_2021_original['ENERGY SOURCE'] == 'Total')
                               ,:].copy()

energy_2021
```

```python
"""
3. Output a CSV file which should contain 3 columns: MONTH, STATE, GENERATION(Mwh)
"""

energy_2021_MSG = energy_2021[['MONTH', 'STATE', 'GENERATION(Mwh)']]
energy_2021_MSG = energy_2021_MSG.groupby(['MONTH', 'STATE']).sum()
energy_2021_MSG = energy_2021_MSG.reset_index()
energy_2021_MSG.to_csv('./output/energy_2021_MSG.csv', index=False)

energy_2021_MSG
```

```python
"""
4. According to the original energy data, select 2021 monthly data based on TYPE OF PRODUCER. \
(Note that ENERGY SOURCE is 'Total')
"""

energy_2021_MT = energy_2021_original.loc[~((energy_2021_original['TYPE OF PRODUCER'] == 'Total Electric Power Industry') |
                                (energy_2021_original['STATE'] == 'US-TOTAL')
                               ) & (energy_2021_original['ENERGY SOURCE'] == 'Total'),:].copy()
energy_2021_MT = energy_2021_MT.reset_index()
energy_2021_MT = energy_2021_MT.drop(columns=['index', 'YEAR', 'MONTH'])

energy_2021_MT
```

```python
"""
5. Calculate the yearly sum of total GENERATION based on TYPE OF PRODUCER and STATE.
6. Use pandas.pivot_table to rearrange the dataframe
7. Output the CSV file which has 3 columns: STATE,TYPE OF PRODUCER, GENERATION(Mwh)
"""

energy_2021_MT_TS = energy_2021_original.loc[~(energy_2021_original['TYPE OF PRODUCER'] == 'Total Electric Power Industry') & 
                                    (energy_2021_original['ENERGY SOURCE'] == 'Total'),:].copy()
energy_2021_MT_TS = energy_2021_MT_TS.reset_index()
energy_2021_MT_TS = energy_2021_MT_TS.drop(columns=['index', 'YEAR', 'MONTH'])
energy_2021_MT_TS = energy_2021_MT.groupby(['STATE', 'TYPE OF PRODUCER']).sum()
energy_2021_MT_TS = energy_2021_MT_TS.reset_index()
energy_2021_MT_TS.to_csv('./output/energy_2021_MT_TS.csv')

energy_2021_MT_TS
```
